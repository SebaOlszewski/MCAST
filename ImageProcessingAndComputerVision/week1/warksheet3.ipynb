```python
BGRimage = cv2.imread('lena.png')
RGBimage = cv2.cvtColor(BGRimage, cv2.COLOR_BGR2RGB)
Rimage = Gimage = Bimage = RGBimage.shape
Rimage = RGBimage.copy()
Gimage = RGBimage.copy()
Bimage = RGBimage.copy()

Rimage[:,:,[1,2]] = 0
Gimage[:,:,[0,2]] = 0
Bimage[:,:,[0,1]] = 0

f, axs = plt.subplots(2,2)
axs[0, 0].plt.imshow(Rimage, "RED")
axs[0, 1].plt.imshow(Gimage, 'Green')
axs[1, 0].plt.imshow(Bimage, 'Blue')
axs[1, 1].plt.imshow(Rimage, 'Combined')
plt.show()



```
