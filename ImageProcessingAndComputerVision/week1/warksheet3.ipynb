```python
#task 2

from plotly.subplots import make_subplots
import plotly.graph_objects as go

BGRimage = cv2.imread('lena.png')
RGBimage = cv2.cvtColor(BGRimage, cv2.COLOR_BGR2RGB)
Rimage = Gimage = Bimage = RGBimage.shape
Rimage = RGBimage.copy()
Gimage = RGBimage.copy()
Bimage = RGBimage.copy()

Rimage[:,:,[1,2]] = 0
Gimage[:,:,[0,2]] = 0
Bimage[:,:,[0,1]] = 0



fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax1.imshow(Rimage)
ax2 = fig.add_subplot(2, 2, 2)
ax2.imshow(Gimage)
ax3 = fig.add_subplot(2, 2, 3)
ax3.imshow(Bimage)
ax4 = fig.add_subplot(2, 2, 4)
ax4.imshow(RGBimage)
Image.fromarray(Rimage).save("LENAred.png")
Image.fromarray(Gimage).save("LENAgreen.png")
Image.fromarray(Bimage).save("LENAblue.png")


plt.show


```
