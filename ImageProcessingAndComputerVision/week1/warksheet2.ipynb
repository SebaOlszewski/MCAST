```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python

#Task1
#loading image as BGR (not RGB standard)
#we are converting it to RGB later on

BGRimage = cv2.imread('lena.png')
RGBimage = cv2.cvtColor(BGRimage, cv2.COLOR_BGR2RGB)
Rimage = Gimage = Bimage = RGBimage.shape
Rimage = RGBimage.copy()
Gimage = RGBimage.copy()
Bimage = RGBimage.copy()

Rimage[:,:,[1,2]] = 0
Gimage[:,:,[0,2]] = 0
Bimage[:,:,[0,1]] = 0

plt.imshow(Rimage)
plt.show()
plt.imshow(Gimage)
plt.show()
plt.imshow(Bimage)
plt.show()
plt.imshow(Rimage + Gimage + Bimage)
plt.show()
```

```python
#task 2
brightening_val = 0
BGRimage = cv2.imread('lena.png')
RGBimage = cv2.cvtColor(BGRimage, cv2.COLOR_BGR2RGB)
plt.imshow(RGBimage)
plt.show()


#RGBimage += brightening_val
RGBimage[RGBimage + brightening_val  >= 255] = 255
RGBimage[RGBimage + brightening_val  < 255] += brightening_val

plt.imshow(RGBimage)
plt.show()
```

```python
#task 3
trim_size = 20
BGRimage = cv2.imread('lena.png')
RGBimage = cv2.cvtColor(BGRimage, cv2.COLOR_BGR2RGB)
print( "Shape before trimming:")
print(RGBimage.shape)


RGBimage = RGBimage[trim_size:, trim_size:, :]
RGBimage = RGBimage[:-trim_size:, :-trim_size:, :]
print( "Shape after trimming:")
print(RGBimage.shape)

plt.imshow(RGBimage)
plt.show()

```
