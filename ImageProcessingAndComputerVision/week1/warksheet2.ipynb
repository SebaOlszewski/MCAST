```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python

#Task1
#loading image as BGR (not RGB standard)
#we are converting it to RGB later on

BGRimage = cv2.imread('lena.png')
RGBimage = cv2.cvtColor(BGRimage, cv2.COLOR_BGR2RGB)
Rimage = Gimage = Bimage = RGBimage.shape
Rimage = RGBimage.copy()
Gimage = RGBimage.copy()
Bimage = RGBimage.copy()

Rimage[:,:,[1,2]] = 0
Gimage[:,:,[0,2]] = 0
Bimage[:,:,[0,1]] = 0

plt.imshow(Rimage)
plt.show()
plt.imshow(Gimage)
plt.show()
plt.imshow(Bimage)
plt.show()
plt.imshow(Rimage + Gimage + Bimage)
plt.show()
```

```python
#task 2
brightening_val = 100
BGRimage = cv2.imread('lena.png')
RGBimage = cv2.cvtColor(BGRimage, cv2.COLOR_BGR2RGB)
plt.imshow(RGBimage)
plt.show()

HSVimage = cv2.cvtColor(RGBimage, cv2.COLOR_RGB2HSV)

h, s, v = cv2.split(HSVimage)

limit = 255 - brightening_val

v[v >= limit] = 255
v[v < limit] += brightening_val

HSVimageoutput = cv2.merge([h, s, v])

RGBimageoutput = cv2.cvtColor(HSVimageoutput, cv2.COLOR_HSV2RGB)

#RGBimage += brightening_val
#RGBimage[:,:,2][RGBimage[:,:,2] + brightening_val  >= 255] = 255
#RGBimage[:,:,2][RGBimage[:,:,2] + brightening_val  < 255] += brightening_val



plt.imshow(RGBimageoutput)
plt.show()
```

```python
#task 3
trim_size = 100
BGRimage = cv2.imread('lena.png')
RGBimage = cv2.cvtColor(BGRimage, cv2.COLOR_BGR2RGB)
print( "Shape before trimming:")
print(RGBimage.shape)


RGBimage = RGBimage[trim_size:-trim_size, trim_size:-trim_size, :]
print( "Shape after trimming:")
print(RGBimage.shape)

plt.imshow(RGBimage)
plt.show()

```
