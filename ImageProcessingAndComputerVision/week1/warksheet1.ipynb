```python
import numpy as np
import matplotlib.pyplot as plt
#import cv2

#task 1
M = np.array(([2, 4, 6, 8],[10, 20, 30, 40],[5, 10, 15, 20]))
print(M)
```

```python
#task 2
print(M[1,:])
```

```python
#task 3
print(M[:,3])
```

```python
#task 4
#matrix M before change
print(M)
print("")

#matrix M after change
for i in range(0, 3):
    for j in range(0, 4):
        if M[i, j] > 9:
            M[i, j] = 0
print(M)
```

```python
#task 5

M2 = np.random.randint(0, 256, size = M.shape, dtype='uint8')
print(M2)
plt.imshow(M2, cmap='gray', vmin=0, vmax=255) #displays the grayscale plane
```

```python
#task 6

M3 = np.zeros([100,200,3], dtype='uint8')
#orange half
M3[:100,:100,:] = [255,128,0]
#blue half
M3[:100,100:,:] = [0,0,255]
#print(M3)
plt.imshow(M3, vmin=0, vmax=255) #displays the checked plane
```

```python
#task 7

M4 = np.zeros([100,200,3], dtype='uint8')
#orange half
M4[:100,:100,:] = [255,128,0]
#blue half
M4[:100,100:,:] = [0,0,255]

#adds the transparency to the RGB array
RGBAarray = np.full(M4.shape[:-1], 100, dtype="uint8")
RGBAarray = np.dstack((M4,RGBAarray))

#plots the array
plt.imshow(RGBAarray, vmin=0, vmax=255)
```
