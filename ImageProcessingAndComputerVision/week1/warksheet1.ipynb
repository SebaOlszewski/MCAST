```python
#task 4
#matrix M before change
print(M)
print("")

#matrix M after change
for i in range(0, 3):
    for j in range(0, 4):
        if M[i, j] > 9:
            M[i, j] = 0
print(M)
```

```python
#does the same thing as the code above ^
M = np.array(([2, 4, 6, 8],[10, 20, 30, 40],[5, 10, 15, 20]))
print(M)
M[M > 9] = 0
print(M)

#using another array for that if we dont want to change the original one
N = M.copy()
N[M > 9] = 0
print(N)

```

```python
#task 6

M3 = np.zeros([100,200,3], dtype='uint8')
#orange half
M3[:100,:100,:] = [255,128,0]
#blue half
M3[:100,100:,:] = [0,0,255]
#print(M3)
plt.imshow(M3, vmin=0, vmax=255) #displays the checked plane

Image.fromarray(M3).save("orange_blue.png")
```

```python
#task 7

M4 = np.zeros([100,200,3], dtype='uint8')
#orange half
M4[:100,:100,:] = [255,128,0]
#blue half
M4[:100,100:,:] = [0,0,255]

#adds the transparency to the RGB array
transparency = 100 #(0, 255)
RGBAarray = np.full(M4.shape[:-1], transparency, dtype="uint8")
RGBAarray = np.dstack((M4,RGBAarray))

#plots the array
plt.imshow(RGBAarray, vmin=0, vmax=255)
```

```python
#task 7 correct way
#we create it with 4 layers, RGBA
M4 = np.zeros([100,200,4], dtype='uint8')
#orange half
M4[:100,:100,:] = [255,128,0, 255]
#blue half
M4[:100,100:,:] = [0,0,255, 255]

for x in range (200):
    M4[:, x, 3] = x
plt.imshow(M4)
plt.show

#plots the array
#plt.imshow(RGBAarray, vmin=0, vmax=255)
```
