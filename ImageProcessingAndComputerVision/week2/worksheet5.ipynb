```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as plt_im
import cv2
import cv2 as cv
```

```python

flamigo_bgr = cv2.imread('flamigo.jpg')
flamigo_rgb = cv2.cvtColor(flamigo_bgr, cv2.COLOR_BGR2RGB)
flamigo_gray1 = cv2.cvtColor(flamigo_bgr, cv2.COLOR_BGR2GRAY)

hsv = cv2.cvtColor(flamigo_bgr, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

s[:] = 0

final_hsv = cv2.merge((h,s,v))
flamigo_gray2 = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)


r, g, b = cv2.split(flamigo_bgr)

GS_value = 0.298936 * r + 0.587043*g + 0.114021*b


plt.title("BGR image")
plt.imshow(flamigo_bgr)
plt.show()
plt.title("RGB image")
plt.imshow(flamigo_rgb)
plt.show()
plt.title("grayscale1 image // converting BGR to GRAY")
plt.imshow(flamigo_gray1, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.title("grayscale2 image // converting to HSV and setting S layer to 0")
plt.imshow(flamigo_gray2, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.title("grayscale3 image // Using the given formula for each pixel")
plt.imshow(GS_value, cmap='gray', vmin=0, vmax=255)
plt.show()
```
