```python
from matplotlib import pyplot as plt
import numpy as np
import math 
import cv2

x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("Sine")
plt.title("Sine wave")
plt.show
```

```python

x = np.array(["Jan","Feb","Mar","April","May","June","July","August","Sept","Oct", "Nov", "Dec"])
y = np.array([95,63, 37, 26, 9, 5, 0, 6, 67, 77, 109, 109])

plt.bar(x,y)
plt.xlabel("Months")
plt.ylabel("mm")
plt.title("Rainfall")
plt.show()
```

```python

x = np.array([22, 87, 5, 78, 32, 25, 83, 51, 75, 83, 89, 91, 25])
bins = [0, 25, 50, 75, 100]

fig, ax = plt.subplots(1,2)

ax[0].hist(x, bins)
ax[0].set_xticks([0,25,50,75,100])   #sets the x axis to make the bars be between the ticks' values
ax[0].set_xlabel("Grades")
ax[0].set_ylabel("No of students")
ax[0].set_title("Exam results")
plt.show()
```

```python
gray_img = cv2.imread('lena.png', 0)


fig, ax = plt.subplots(1,2)

flattened_img = gray_img.ravel()  #ravel() function flattenes a 2d array to 1d array
                                    #in this scenario converts the image to 1d array

ax[0].imshow(gray_img, cmap = "grey")

ax[1].hist(flattened_img, 256)  #sets the x axis to make the bars be between the ticks' values
ax[1].set_xlabel("Intensity")
ax[1].set_ylabel("No of pixels")
ax[1].set_title("Image histogram")
plt.show()
```

```python
gray_img = cv2.imread('lena.png', 0)

plt.imshow(gray_img, cmap = "grey")
plt.show()

hist = cv2.calcHist(gray_img, [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()

gray_norm = cv2.equalizeHist(gray_img)
plt.imshow(gray_norm, cmap = "grey")
plt.show()
```
